```python
# ============================================================
# FINAL STABLE VERSION (KAGGLE TF COMPATIBLE)
# DenseNet201 + 2x T4 + NO CollectiveReduce errors
# ============================================================

import os, numpy as np, tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import DenseNet201
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# CLEAN START
# -------------------------
tf.keras.backend.clear_session()
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

# -------------------------
# PATHS
# -------------------------
train_dir = '/kaggle/input/skin-mal-ben-bcc-balanced/ISIC_BALANCED/train'
val_dir   = '/kaggle/input/skin-mal-ben-bcc-balanced/ISIC_BALANCED/val'
test_dir  = '/kaggle/input/skin-mal-ben-bcc-balanced/ISIC_BALANCED/test'

# -------------------------
# CONFIG
# -------------------------
IMG_SIZE = 224
BATCH_SIZE = 16   # PER GPU → GLOBAL = 32
EPOCHS = 30
NUM_CLASSES = 3
CLASS_NAMES = ['BCC', 'benign', 'malignant']

# -------------------------
# MULTI-GPU STRATEGY
# -------------------------
strategy = tf.distribute.MirroredStrategy()
print("GPUs in use:", strategy.num_replicas_in_sync)

# -------------------------
# DATA PIPELINE (FIXED FOR KAGGLE)
# -------------------------
def make_dataset(path, shuffle=False):
    ds = tf.keras.utils.image_dataset_from_directory(
        path,
        image_size=(IMG_SIZE, IMG_SIZE),
        batch_size=None,              # 🔥 NO batching here
        label_mode='categorical',
        shuffle=shuffle
    )
    ds = ds.map(
        lambda x, y: (tf.cast(x, tf.float32) / 255.0, y),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    ds = ds.batch(BATCH_SIZE, drop_remainder=True)  # 🔥 ONLY batching
    return ds.prefetch(tf.data.AUTOTUNE)

train_ds = make_dataset(train_dir, shuffle=True)
val_ds   = make_dataset(val_dir)
test_ds  = make_dataset(test_dir)

# -------------------------
# FOCAL LOSS
# -------------------------
def focal_loss(gamma=2.0, alpha=0.25):
    def loss(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1 - 1e-7)
        ce = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.pow(1 - y_pred, gamma)
        return tf.reduce_sum(weight * ce, axis=1)
    return loss

# -------------------------
# MODEL
# -------------------------
with strategy.scope():
    base = DenseNet201(
        include_top=False,
        weights='imagenet',
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )

    # Freeze backbone (stable)
    for layer in base.layers[:-60]:
        layer.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    output = layers.Dense(NUM_CLASSES, activation='softmax')(x)

    model = models.Model(base.input, output)

    model.compile(
        optimizer=optimizers.Adam(1e-4),
        loss=focal_loss(),
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc', multi_label=True)
        ]
    )

model.summary()

# -------------------------
# TRAIN
# -------------------------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)

# -------------------------
# EVALUATE
# -------------------------
pred_probs = model.predict(test_ds)
y_true = np.concatenate([y.numpy() for _, y in test_ds])
y_true = np.argmax(y_true, axis=1)
y_pred = np.argmax(pred_probs, axis=1)

print("\nCLASSIFICATION REPORT\n")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES))

print("Macro AUC-ROC:", roc_auc_score(y_true, pred_probs, multi_class='ovr'))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=CLASS_NAMES,
            yticklabels=CLASS_NAMES)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
# ============================================================
# DENSENET — FULL METRICS CALCULATION
# ============================================================

import numpy as np
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    accuracy_score,
    precision_recall_fscore_support
)
import seaborn as sns
import matplotlib.pyplot as plt

CLASS_NAMES = ['BCC', 'benign', 'malignant']

# -------------------------
# GET PREDICTIONS
# -------------------------
y_true = []
y_prob = []

for x, y in test_ds:
    y_true.append(y.numpy())
    y_prob.append(model.predict(x, verbose=0))

y_true = np.concatenate(y_true)
y_prob = np.concatenate(y_prob)

y_true_cls = np.argmax(y_true, axis=1)
y_pred_cls = np.argmax(y_prob, axis=1)

# -------------------------
# BASIC METRICS
# -------------------------
accuracy = accuracy_score(y_true_cls, y_pred_cls)

precision, recall, f1, support = precision_recall_fscore_support(
    y_true_cls,
    y_pred_cls,
    average=None
)

macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(
    y_true_cls,
    y_pred_cls,
    average='macro'
)

weighted_precision, weighted_recall, weighted_f1, _ = precision_recall_fscore_support(
    y_true_cls,
    y_pred_cls,
    average='weighted'
)

# -------------------------
# AUC-ROC
# -------------------------
auc_per_class = roc_auc_score(
    y_true,
    y_prob,
    multi_class='ovr',
    average=None
)

macro_auc = roc_auc_score(
    y_true,
    y_prob,
    multi_class='ovr',
    average='macro'
)

weighted_auc = roc_auc_score(
    y_true,
    y_prob,
    multi_class='ovr',
    average='weighted'
)

# -------------------------
# PRINT RESULTS
# -------------------------
print("\n=== OVERALL METRICS ===")
print(f"Accuracy: {accuracy:.4f}")
print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall:    {macro_recall:.4f}")
print(f"Macro F1-score:  {macro_f1:.4f}")
print(f"Macro AUC-ROC:   {macro_auc:.4f}")
print(f"Weighted AUC:    {weighted_auc:.4f}")

print("\n=== PER-CLASS METRICS ===")
for i, cls in enumerate(CLASS_NAMES):
    print(
        f"{cls:<10} | "
        f"Precision: {precision[i]:.3f} | "
        f"Recall: {recall[i]:.3f} | "
        f"F1: {f1[i]:.3f} | "
        f"AUC: {auc_per_class[i]:.3f} | "
        f"Support: {support[i]}"
    )

# -------------------------
# CLASSIFICATION REPORT
# -------------------------
print("\n=== CLASSIFICATION REPORT ===\n")
print(classification_report(
    y_true_cls,
    y_pred_cls,
    target_names=CLASS_NAMES
))

# -------------------------
# CONFUSION MATRIX
# -------------------------
cm = confusion_matrix(y_true_cls, y_pred_cls)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=CLASS_NAMES,
    yticklabels=CLASS_NAMES
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("DenseNet Confusion Matrix")
plt.show()

# -------------------------
# IMPORTANT MEDICAL METRIC
# -------------------------
malignant_index = CLASS_NAMES.index('malignant')
print(f"\n🔥 Malignant Recall (Sensitivity): {recall[malignant_index]:.4f}")
```
